```python
import pandas as pd
import matplotlib.pyplot as plt

url_dados='https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa= {'droga' : 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
import seaborn as sns
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto' , data=dados.query('composto in @cod_compostos'), order= cod_compostos)
ax.set_title('Top 5 compostos', fontdict={'fontsize':18, 'fontweight': 'bold'})
ax.set_xlabel('Compostos', fontdict={'fontsize':16})
ax.set_ylabel('Quantidade', fontdict={'fontsize':16})
ax.tick_params(labelsize=15)
plt.show()
```
